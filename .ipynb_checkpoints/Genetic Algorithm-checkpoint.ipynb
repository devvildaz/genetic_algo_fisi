## Problema en base decimal

```python
import random
import math
```

```python
queens = 8
def generate_sample(n_sample):
    sample = []
    for _ in range(n_sample):
        sample.append([int(random.random()*queens) for i in range(queens)])

    return sample
```

```python
# function test
generate_sample(4)
```

```python
def fitness(chromosome):
    score = 0
    for gen_col, gen_row in enumerate(chromosome):
        q_attacked = 0
        for column, row in enumerate(chromosome):
            if gen_col == column:
                continue
            x = gen_col - column
            y = gen_row - row
            """
            res = math.pow(x,2) + math.pow(y,2)
            res = math.sqrt(res)
            """
            if row == gen_row:
                q_attacked += 1
            elif x**2 == y**2:
                q_attacked += 1
            """
            print(res)
            if math.floor(res) == res or math.ceil(res) == res:
                q_attacked += 1
            """
        score += q_attacked
    return score
```

```python
fitness([2,3,4,1,6,7,1,5])
```

```python
def roulette_wheel(sample):
    n = len(sample)
    arr_fit = [(chromosome,fitness(chromosome)) for chromosome in sample]
    accum = sum(i[1] for i in arr_fit)
    prob = [(tupl[0],tupl[1]/accum) for tupl in arr_fit]
    prob.sort(key=lambda x:x[1])
    roulette = []
    for idx, item in enumerate(prob):
        if idx == 0:
            roulette.append((item[0],item[1]))
        else:
            roulette.append((item[0], roulette[idx-1][1] + item[1]))
    
    new_sample = []
    for i in range(n//2):
        dice_score = random.random()
        for item in roulette:
            if dice_score < item[1]:
                new_sample.append(item[0])
                new_sample.append(item[0])
                break
    return new_sample
    
    
```

```python
roulette_wheel(generate_sample(4))
```
